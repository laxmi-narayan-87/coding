```python
/import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

xpts=np.array([0,10])
ypts=np.array([0,125])
plt.plot(xpts,ypts)
plt.show()


data={"Name": ['ABC','DEF','GHI','JKL'],
      "Roll no.": [20,30,40,50],
      "Height":[40,70,60,30]
        }
df=pd.DataFrame(data)
plt.plot(df['Roll no.'],df['Height'])
plt.show()
plt.scatter(df['Roll no.'],df['Height'])
plt.show()
plt.bar(df['Roll no.'],df['Height'])
plt.show()
plt.hist(df['Height'])
plt.show()
plt.pie(df['Height'],labels=df['Name'])
plt.show()
plt.barh(df['Name'],df['Height'])
plt.show()
plt.boxplot(df['Height'])
plt.show()

a=np.arange(5)
b=[2,4,6,8,10]
c=[5,6,7,8,9]
fig=plt.figure()
ax=plt.subplot()
ax.plot(a,b,'k--',label='b')
ax.plot(a,c,'k:',label='c')
ax.legend(loc='upper right')
plt.show()
```
