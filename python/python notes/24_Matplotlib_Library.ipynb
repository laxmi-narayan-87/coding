# Matplotlib

Matplotlib is a powerful Python library used for creating visualizations such as plots, graphs, and charts. It's widely used in data science, machine learning, and scientific computing for understanding and presenting data in an easy-to-understand format.

# What is Matplotlib?

Matplotlib is a 2D plotting library for Python. It provides tools to create a variety of plots like:

Line plots

Bar plots

Histograms

Scatter plots

Pie charts

Heatmaps

It’s flexible and allows for both simple and complex visualizations. You can think of it as a way to draw pictures using data points.

# Why Use Matplotlib

Data visualization helps you to understand the data better.

It can show trends, patterns, or anomalies in datasets.

You can create customized plots for reports and presentations.

It supports different formats, like PNG, JPG, and PDF, for saving your plots.

# Main Components of Matplotlib

Matplotlib is highly structured but also easy to use once you understand its components

a) Pyplot

The pyplot module is the most used part of Matplotlib. It is often used to quickly plot data with a few lines of code. Think of pyplot as the artist that helps you draw your visualizations. Most of the time, you’ll see import matplotlib.pyplot as plt.

b) Figure

A figure is the outermost container for a plot. It can contain multiple plots (also called subplots). You can think of a figure as a blank canvas where the plots are drawn.

c) Axes

The axes are the actual plotting area where data is drawn. Each plot has its own axes, which contain x-axis and y-axis with labels, ticks, and grids.

# Advantages of Matplotlib

Simplicity: Pyplot makes it very simple to create basic plots with minimal code.

Highly Customizable: You can control every element of the plot (title, labels, colors, grids, etc.).

Wide Variety of Plots: Matplotlib supports many different types of plots and can handle complex visualizations.

Integration with Other Libraries: Works well with pandas and NumPy for data manipulation and plotting.

Publication-Quality: You can create high-quality plots suitable for publication or presentation.

# Disadvantages

Steeper Learning Curve for Complex Plots: While simple plots are easy to make, more complex plots and customization require a deeper understanding.

Limited Interactivity: Compared to some other visualization libraries like Plotly, Matplotlib has limited interactivity.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\Users\\LENOVO\\Desktop\\IPL_Ball-by-Ball_2008-2020.csv")
df.head()
```

```python
Total_Wickets_by_Bowler = df.groupby("bowler")["is_wicket"].sum().sort_values(ascending = False).reset_index().head()
Total_Wickets_by_Bowler
```

```python
plt.figure(figsize=(8,4))
plt.bar(Total_Wickets_by_Bowler["bowler"],Total_Wickets_by_Bowler["is_wicket"])
plt.title("Most Wicket Taker Bolwer",fontsize = 20,color = "g",loc = "center")
plt.xlabel("Player Name",fontsize = 20,color = "g",loc = "center")
plt.ylabel("Number of Wickets",fontsize = 20,color = "g",loc = "center")
plt.xticks(rotation = 45)
plt.yticks()
plt.show()
```

```python
# Create a new figure with specified size (width=8, height=4) and background color
plt.figure(
    figsize=(8, 4),  # Size of the figure (width, height)
    dpi=100,  # Dots per inch (controls resolution)
    facecolor="white",  # Background color of the figure
    edgecolor="black",  # Color of the figure edge (border)
    linewidth=2  # Thickness of the edge (border)
)

# Plot a bar chart with the bowler names on x-axis and wickets on y-axis
plt.bar(
    Total_Wickets_by_Bowler["bowler"],  # x-axis data (names of bowlers)
    Total_Wickets_by_Bowler["is_wicket"],  # y-axis data (number of wickets taken)
    color=["#f72585", "#7209b7", "yellow", "cyan", "magenta"],  # color for each bar, list of colors provided
    edgecolor="blue",  # Color of the edge around each bar
    linewidth=2,  # Width of the edge around each bar
    linestyle="--",  # Style of the edge, dashed line here
    alpha=0.9,  # Transparency of the bars, closer to 1 means less transparent
    width=0.8,  # Width of the bars (default is 0.8)
    align="center"  # Alignment of the bars ('center', 'edge')
)

# Add a title to the plot
plt.title(
    label="Most Wicket Taker Bowler",  # Title text
    fontsize=20,  # Font size of the title
    color="black",  # Color of the title text
    loc="center",  # Location of the title ('center', 'left', 'right')
    pad=10,  # Padding between the title and the plot (in points)
    fontweight='bold',  # Weight of the font (e.g., 'normal', 'bold')
    backgroundcolor='lightgray'  # Background color behind the title text
)

# Label the x-axis with "Player Name"
plt.xlabel(
    xlabel="Player Name",  # Label text for x-axis
    fontsize=20,  # Font size of the x-axis label
    color="black",  # Color of the x-axis label text
    loc="center",  # Location of the x-axis label ('center', 'left', 'right')
    labelpad=10,  # Padding between the x-axis label and the axis
    fontweight='bold',  # Weight of the font for the x-axis label
    backgroundcolor='lightgray'  # Background color behind the x-axis label text
)

# Label the y-axis with "Number of Wickets"
plt.ylabel(
    ylabel="Number of Wickets",  # Label text for y-axis
    fontsize=20,  # Font size of the y-axis label
    color="black",  # Color of the y-axis label text
    loc="center",  # Location of the y-axis label ('center', 'top', 'bottom')
    labelpad=10,  # Padding between the y-axis label and the axis
    fontweight='bold',  # Weight of the font for the y-axis label
    backgroundcolor='lightgray'  # Background color behind the y-axis label text
)

# Rotate the x-axis labels by 45 degrees for better visibility
plt.xticks(
    ticks=None,  # Specifies the locations of the ticks (default: None, meaning use default locations)
    labels=None,  # Labels for the ticks (default: None, meaning use the original labels from the data)
    rotation=45,  # Rotation angle for x-axis labels
    fontsize=12,  # Font size of the x-axis tick labels
    color="blue",  # Color of the x-axis tick labels
    fontweight='bold'  # Font weight for x-axis tick labels
)

# Keep the y-axis labels unchanged (this can be used to format the labels, but it's not necessary here)
plt.yticks(
    ticks=None,  # Specifies the locations of the y-axis ticks (default: None)
    labels=None,  # Labels for the y-axis ticks (default: None, meaning use the original labels from the data)
    fontsize=12,  # Font size of the y-axis tick labels
    color="blue",  # Color of the y-axis tick labels
    fontweight='bold'  # Font weight for y-axis tick labels
)

# Disable the gridlines in the plot
plt.grid(
    visible=False,      # Controls whether grid is on or off (False = no grid, True = grid is shown)
    which='both',       # Which grid lines to apply the settings ('both', 'major', 'minor')
    axis='both',        # Which axis to apply the grid (default is 'both' for x and y axes)
    color='gray',       # Color of the grid lines (if they were shown)
    linestyle='-',      # Style of the grid lines (if they were shown)
    linewidth=0.5,      # Width of the grid lines (if they were shown)
    alpha=0.7           # Transparency level of the grid lines (if they were shown)
)

# Display the plot
plt.show()  # Renders the plot to the screen
```

```python
# Data for plotting
x = [2, 4, 6, 8, 10]
y = [5, 10, 15, 5, 10]

# Create a new figure with custom size and background color
plt.figure(
    figsize=(10, 6),  # Custom figure size (width, height)
    dpi=100,          # Resolution of the figure
    facecolor='white' # Background color
)

# Plot the line with specified style
plt.plot(
    x, y,
    color="g",             # Green line color
    linewidth=3,           # Thickness of the line
    linestyle="--",        # Dashed line style
    marker="<",            # Left-facing triangle markers
    markerfacecolor='red', # Red fill color for markers
    markeredgewidth=2,     # Border thickness of markers
    markeredgecolor='black', # Black border color for markers
    markersize=10          # Size of the markers
)

# Add a title to the plot
plt.title(
    "Sample Line Chart",        # Title of the plot
    fontsize=20,                # Font size of the title
    color="blue",               # Title color
    loc="center",               # Title alignment (centered)
    pad=20,                     # Padding between title and plot
    fontweight='bold'           # Bold title font
)

# Label the x-axis
plt.xlabel(
    "X Values",                 # X-axis label text
    fontsize=15,                # Font size of the label
    color="darkred",            # Label color
    labelpad=10,                # Padding between the label and axis
    fontweight='bold'           # Bold font for label
)

# Label the y-axis
plt.ylabel(
    "Y Values",                 # Y-axis label text
    fontsize=15,                # Font size of the label
    color="darkred",            # Label color
    labelpad=10,                # Padding between the label and axis
    fontweight='bold'           # Bold font for label
)

# Add gridlines for better visibility
plt.grid(
    visible=True,               # Turn on gridlines
    which='major',              # Show major gridlines
    axis='both',                # Apply gridlines to both axes
    color='gray',               # Gridline color
    linestyle='-',              # Solid line style for grid
    linewidth=0.5,              # Gridline width
    alpha=0.7                   # Transparency of the grid
)

# Set custom ticks on x-axis and y-axis
plt.xticks(
    ticks=x,                    # Custom ticks on the x-axis
    fontsize=12,                # Font size of x-axis labels
    color="black",              # Color of the x-axis labels
    fontweight='bold'           # Bold font for labels
)
plt.yticks(
    fontsize=12,                # Font size of y-axis labels
    color="black",              # Color of the y-axis labels
    fontweight='bold'           # Bold font for labels
)

# Add a background color for the plot area
plt.gca().set_facecolor('#f0f0f0')  # Light gray background for the plot area

# Display the plot
plt.show()
```

```python
# Data for scatter plot
a = [10, 20, 30, 40, 50, 60]
b = [45, 20, 60, 50, 30, 80]

# Create a new figure with a custom size
plt.figure(
    figsize=(10, 6),     # Custom figure size (width, height)
    dpi=100,             # Resolution of the figure
    facecolor='white'    # Background color of the figure
)

# Plot the scatter plot with detailed styling
plt.scatter(
    a,                      # Data for x-axis
    b,                      # Data for y-axis
    marker="*",             # Star-shaped marker for data points
    s=150,                  # Size of the markers (points)
    color="purple",         # Marker color
    edgecolor='black',      # Edge color of the markers to make them more distinct
    linewidth=1,            # Line width of marker edges
    alpha=0.8               # Transparency level (1 is opaque, 0 is transparent)
)

# Add grid with custom styling
plt.grid(
    linestyle="--",         # Dashed lines for the grid
    alpha=0.3,              # Transparency of the grid lines (lighter)
    color="grey",           # Grid line color
    linewidth=0.7           # Line width of the grid
)

# Set labels for the axes
plt.xlabel(
    "X Values",             # Label for x-axis
    fontsize=14,            # Font size for x-axis label
    color="blue",           # Color of the x-axis label
    fontweight='bold'       # Bold font for x-axis label
)

plt.ylabel(
    "Y Values",             # Label for y-axis
    fontsize=14,            # Font size for y-axis label
    color="blue",           # Color of the y-axis label
    fontweight='bold'       # Bold font for y-axis label
)

# Set title for the plot
plt.title(
    "Scatter Plot of X vs Y",  # Title of the chart
    fontsize=20,               # Font size of the title
    color="darkred",           # Color of the title text
    pad=20,                    # Padding between the title and the plot
    fontweight='bold'          # Bold font for the title
)

# Show the plot with the custom settings
plt.show()
```

```python
# Creating a sample dataset
np.random.seed(42)
data = {
    'Age': np.random.randint(18, 65, 50),
    'Income': np.random.randint(30000, 120000, 50),  # Income in dollars
    'Years_of_Education': np.random.randint(10, 20, 50),
    'Work_Experience': np.random.randint(1, 40, 50),
    'Health_Score': np.random.uniform(50, 100, 50),  # Health score out of 100
    'Loan_Amount': np.random.randint(5000, 50000, 50),
    'Credit_Score': np.random.randint(300, 850, 50),  # Credit score range
    'Satisfaction_Score': np.random.randint(1, 10, 50)  # Customer satisfaction (1-10)
}

df = pd.DataFrame(data)


df.head()
```

```python
correlation_matrix = df.corr(numeric_only = True)
correlation_matrix 
```

```python
cmap_names = plt.colormaps()
print(cmap_names)
```

```python
plt.style.use("dark_background")

plt.figure(figsize = (12,8))

plt.imshow(correlation_matrix ,cmap = "viridis")

plt.colorbar()

plt.xticks(range(len(correlation_matrix)),correlation_matrix.columns,rotation = 45)
plt.yticks(range(len(correlation_matrix)),correlation_matrix.columns)

for i in range(len(correlation_matrix)):
    for j in range(len(correlation_matrix)):
        plt.text(j,i,f'{correlation_matrix.iloc[i,j]:.2f}',ha = "center",va = "center",color = "black")

plt.title("Correlation Matrix")

plt.grid(False)
plt.show()
```

```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create a boxplot with customizations
box = ax.boxplot(df['Income'], 
                 patch_artist=True,  # Allows us to color the box
                 boxprops=dict(facecolor='lightblue', color='blue'),  # Box color
                 whiskerprops=dict(color='blue', linewidth=2),  # Whisker color and width
                 capprops=dict(color='blue', linewidth=2),  # Caps color and width
                 medianprops=dict(color='red', linewidth=2),  # Median line color and width
                 flierprops=dict(marker='o', markerfacecolor='orange', markersize=8, linestyle='none'))  # Outlier customization

# Customize the plot
ax.set_title('Income Distribution', fontsize=16, fontweight='bold', color='darkblue')  # Adding a title
ax.set_ylabel('Income', fontsize=14)  # Label for y-axis
ax.set_xticks([1])  # Custom x-tick
ax.set_xticklabels(['Income'], fontsize=12)  # Label for the x-axis

# Set background grid
ax.grid(False)

# Show the plot
plt.tight_layout()
plt.show()
```
